```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img_path = "0005.png"
img = cv2.imread(img_path,cv2.IMREAD_UNCHANGED)
plt.imshow(img[:,:,::-1])
plt.show()
```

```python
blur = img.copy()
blur.astype(dtype=np.uint8)
for item in range(10):
    #blur = cv2.blur(blur,(5,5))
    blur = cv2.GaussianBlur(blur,(5,5),0)
#blur = cv2.blur(img,(5,5))
plt.imshow(blur[:,:,::-1])
plt.show()
```

```python
gray = cv2.cvtColor(blur, cv2.COLOR_BGR2GRAY)
plt.imshow(gray,cmap='gray')
plt.show()
```

```python
adapt_thres = cv2.adaptiveThreshold(gray,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,\
            cv2.THRESH_BINARY,15,2)
adapt_thres = 255 - adapt_thres
adapt_thres.astype(np.uint16)
plt.imshow(adapt_thres,cmap='gray')
plt.show()
```

```python
ret,thresh2 = cv2.threshold(gray,150,220,cv2.THRESH_BINARY)
thresh2.astype(np.uint16)
thresh2 = 255-thresh2
plt.imshow(thresh2,cmap='gray')
plt.show()
```

```python
adapt_thres2 = cv2.adaptiveThreshold(gray,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,\
            cv2.THRESH_BINARY,21,5)
adapt_thres2.astype(np.uint16)
plt.imshow(255-adapt_thres2,cmap='gray')
plt.show()
```

```python

total_img = adapt_thres + thresh2 + adapt_thres2
total_img.astype(np.float)
plt.imshow(total_img,cmap='gray')
plt.show()
```
