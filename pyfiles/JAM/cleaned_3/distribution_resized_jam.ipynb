```python
import glob
import cv2
import os
from matplotlib import pyplot as plt
#print(glob.glob('*.png'))
png_files = glob.glob('resized_JAM/*.png')
#png_files = glob.glob('*.png')
```

```python
index = {}
str_ = "0123456789=+-"
for i in str_:
    index[i]=0
index
```

```python
for item in png_files:
    ch = str(item.split('/')[1])[0]
    index[ch] += 1
    #print(ch)
    #break
```

```python
index
```

```python
#index
from matplotlib.pyplot import figure

figure(num=None, figsize=(12, 5), dpi=100, facecolor='w', edgecolor='k')
plt.bar(index.keys(), index.values(), width=0.9, color='#1f47e6')
plt.title('images-1L-processed dataset frequency ')
plt.xlabel("Characters ", fontsize=10)  
plt.ylabel("Frequency  ", fontsize=10)
#plt.xticks(fontsize=14)  
#plt.yticks(fontsize=14)
plt.show()
```
