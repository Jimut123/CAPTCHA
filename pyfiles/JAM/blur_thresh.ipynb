```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
#img_path = "0-0.png"
#img_path = "1+7.png"
#img_path = "2-1.png"
img_path = "3-1.png"
img = cv2.imread(img_path,cv2.IMREAD_GRAYSCALE)
plt.imshow(img,cmap='gray')
plt.show()
```

```python
import plotly.graph_objects as go
fig = go.Figure(data=[
    go.Surface(z=img[...,::-1,:])])
fig.show()
```

```python
blur = img.copy()
blur.astype(dtype=np.uint8)
for item in range(2):
    blur = cv2.GaussianBlur(blur,(3,3),0)
    #blur = cv2.blur(blur,(5,5))
plt.imshow(blur,cmap='gray')
plt.show()
```

```python
import plotly.graph_objects as go
fig = go.Figure(data=[
    go.Surface(z=blur[...,::-1,:])])
fig.show()
```

```python
blur_thesh = cv2.threshold(blur,120,200,cv2.THRESH_BINARY)
plt.imshow(blur_thesh[1],cmap='gray')
plt.show()
```

```python
final_img = blur_thesh[1].copy()
import plotly.graph_objects as go
fig = go.Figure(data=[
    go.Surface(z=final_img[...,::-1,:])])
fig.show()
```
