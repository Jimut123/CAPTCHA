```python
from PIL import Image
img = Image.open('captcha_1.jpeg')
```

```python
pixels = img.load() # create the pixel map
for i in range(img.size[0]): # for every pixel:
    for j in range(img.size[1]):
        r,g,b = pixels[i,j]
        if r+g+b <= 80:
            pixels[i,j] = (0, 0 ,0)
        else:
            pixels[i,j] = (255,255,255)
```
